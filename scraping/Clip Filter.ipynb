```python
from transformers import CLIPProcessor, CLIPModel
from selenium.webdriver.chrome.service import Service
from PIL import Image
import cv2
import os
import shutil
from datetime import datetime


model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
counter = 0
path = 'C:\\Users\\AIS\\Desktop\\filter clip\\72'
list_folder = os.listdir(path)
for folder in list_folder:
    print(f'TTTTTTTTTTTT        {folder}           TTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTTT')
    list_files = os.listdir(f'{path}\\{folder}')
    for item in list_files:
        if item.startswith('image'):
            number = item.split('_')[1].split('.')[0]
            print(item)
            with Image.open(f'{path}\\{folder}\\{item}') as im:
                try:
                    inputs = processor(text=['aircrafts','people','ship','train'], images=im, return_tensors="pt", padding=True)
                except Exception as e:
                    print(e)
                    continue
            try:        
                outputs = model(**inputs)
                logits_per_image = outputs.logits_per_image  # this is the image-text similarity score
                probs = logits_per_image.softmax(dim=1)  # we can take the softmax to get the label probabilities
                if probs[0][0] > 0.75:
                    counter += 1
                else:
                    os.remove(f'{path}\\{folder}\\{item}')
                    os.remove(f'{path}\\{folder}\\Caption_{number}.txt')
                    os.remove(f'{path}\\{folder}\\Paragragh_{number}.txt')
                    os.remove(f'{path}\\{folder}\\Link_{number}.txt')
            except:
                print('becuse of the AttributeError deleted')
                os.remove(f'{path}\\{folder}\\{item}')
                os.remove(f'{path}\\{folder}\\Caption_{number}.txt')
                os.remove(f'{path}\\{folder}\\Paragragh_{number}.txt')
                os.remove(f'{path}\\{folder}\\Link_{number}.txt')
    list_files = os.listdir(f'{path}\\{folder}')
    condition = True
    for item in list_files:
        if item.startswith('image'): 
            condition = False
            break
    if condition:
        shutil.rmtree(f'{path}\\{folder}')
    
print('Total image=', counter)
print(datetime.now().time())
```
