```python
import pandas as pd
from io import BytesIO
import base64
from PIL import Image
import cv2
import os
```

```python
def b64(im_path):
    fn = im_path
    #img = Image.open(fn)
    with Image.open(fn) as img:
        img_buffer = BytesIO()
        img.save(img_buffer, format=img.format)
    byte_data = img_buffer.getvalue()
    base64_str = str(base64.b64encode(byte_data))[2:-1]
    return base64_str
```
